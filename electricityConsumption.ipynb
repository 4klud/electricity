```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns
color = sns.color_palette()
%matplotlib inline
```

```python
df = pd.ExcelFile('/Users/taylormade/Desktop/DataScience/ElectricityConsumption.xlsx')
df = df.parse('Sheet1', names = ['Area', 'DomesticSales', 'DomesticMeterNum'])
df
```
